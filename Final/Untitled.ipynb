```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns 
import statsmodels.api as sm
%matplotlib inline
```

```python
boros = gpd.read_file(r'C:/Users/ALIENWARE/Desktop/Urban Data Informatics/Final/boros.geojson')
```

```python
bk = boros[boros['BoroName'] == 'Brooklyn']

# Optionally, plot the filtered GeoDataFrame to check the Manhattan boundaries
boros.plot()
plt.title("Map of Brooklyn")
plt.show()
```

```python
north_brooklyn_neighborhoods = [
    "Williamsburg", "Greenpoint", "Bushwick North", "Bushwick South", 
    "Bedford", "Clinton Hill", "Fort Greene", "Stuyvesant Heights",
    "North Side-South Side", "East Williamsburg"
]
```

```python
bk_north = boros[boros['NTAName'].isin(north_brooklyn_neighborhoods)]

# Optionally, plot the filtered GeoDataFrame to check the selection 
bk_north.plot()
plt.title("Map of Brooklyn")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))


# Plot the map
bk_north.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.8)

# Add labels for each neighborhood
for idx, row in bk_north.iterrows():
    # Get the centroid of each neighborhood
    centroid = row['geometry'].centroid
    # Add the neighborhood name as a label
    ax.text(centroid.x, centroid.y, row['NTAName'], fontsize=8, ha='center', color='black')

# Title and show the map
ax.set_title("Map of Brooklyn with Neighborhood Names", fontsize=16)
plt.savefig("Map of Brooklyn with Neighborhood Names", dpi=300, transparent=True)
plt.show()
```
